```python
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
df = pd.read_csv('/Users/monicaclairew/Documents/ESG Project/Test Data/PredictLit.Rate.csv')
df1 = pd.read_csv('/Users/monicaclairew/Documents/ESG Project/Test Data/Book1.csv')
```

```python
LitX = df['GDP.a'].values[:,np.newaxis]
Lity = df['Lit.Rate'].values
Litx2 = df1['GDP'].values[:,np.newaxis]
```

```python
Lit_X_train = LitX[:-10]
Lit_X_test = LitX[-11:]
```

```python
Lit_y_train = Lity[:-10]
Lit_y_test = Lity[-11:]
```

```python
model = linear_model.LinearRegression()
model.fit(Lit_X_train, Lit_y_train)
```

```python
Lit_y_pred = model.predict(Litx2)
print(Lit_y_pred)
```

```python
# The coefficients
print("Coefficients: \n", model.coef_)

# The mean squared error
print("Mean squared error test_data/pred: %.2f" % mean_squared_error(Lit_y_test, Lit_y_pred))

# The mean absolute error
print("Mean absolute error test_data/pred: %.2f" % mean_absolute_error(Lit_y_test, Lit_y_pred))

# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination test_data/pred: %.2f" % r2_score(Lit_y_test, Lit_y_pred))
```

```python
#reg plot of y-predicted
plt.scatter(Lit_X_train, Lit_y_train, color="black")
plt.plot(Lit_X_test, Lit_y_pred, color="blue", linewidth=3)
plt.title('y-predicted')
```

```python
#reg plot of y-true
plt.scatter(Lit_X_train, Lit_y_train, color="black")
plt.plot(Lit_X_test, Lit_y_test, color="blue", linewidth=3)
plt.title('y-true')
```

```python
#true values vs predicted values
plt.scatter(Lit_y_test, Lit_y_pred)
plt.xlabel('y_test')
plt.ylabel('y_pred')
```
