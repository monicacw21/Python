```python
import yfinance as yF
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib as plt
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
start = datetime.datetime(2014,1,2)
end = datetime.datetime(2022,5,4)
```

```python
dfJPST = yF.download("JPST", start=start, end=end)
dfAGG = yF.download("AGG", start=start, end=end)
dfDIA = yF.download("DIA", start=start, end=end)
dfQQQ = yF.download("QQQ", start=start, end=end)
dfSPY = yF.download("SPY", start=start, end=end)
dfIEFA = yF.download("IEFA", start=start, end=end)
dfIEMG = yF.download("IEMG", start=start, end=end)
dfACWV = yF.download("ACWV", start=start, end=end)
dfVSS = yF.download("VSS", start=start, end=end)
dfVB = yF.download("VB", start=start, end=end)
dfNFRA = yF.download("NFRA", start=start, end=end)
dfIYW = yF.download("IYW", start=start, end=end)
dfUSRT = yF.download("USRT", start=start, end=end)
dfIAGG = yF.download("IAGG", start=start, end=end)
dfEMB = yF.download("EMB", start=start, end=end)
dfHEFA = yF.download("HEFA", start=start, end=end)
dfHEEM = yF.download("HEEM", start=start, end=end)
dfGAL = yF.download("GAL", start=start, end=end)
```

```python
dfJPST.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfAGG.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfDIA.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfQQQ.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfSPY.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfIEFA.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfIEMG.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfACWV.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfVSS.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfVB.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfNFRA.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfIYW.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfUSRT.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfIAGG.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfEMB.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfHEFA.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfHEEM.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
dfGAL.drop(['Open','High','Low','Adj Close','Volume'], axis=1, inplace=True)
```

```python
d1 = pd.merge(dfJPST, dfAGG, how='outer', on='Date', suffixes=('_JPST', '_AGG'))
d2 = pd.merge(d1, dfDIA, how='outer', on='Date', suffixes=('_AGG', '_DIA'))
d3 = pd.merge(d2, dfQQQ, how='outer', on='Date', suffixes=('_DIA', '_QQQ'))
d4 = pd.merge(d3, dfSPY, how='outer', on='Date', suffixes=('_QQQ', '_SPY'))
d5 = pd.merge(d4, dfIEFA, how='outer', on='Date', suffixes=('_SPY', '_IEFA'))
d6 = pd.merge(d5, dfIEMG, how='outer', on='Date', suffixes=('_IEFA', '_IEMG'))
d7 = pd.merge(d6, dfACWV, how='outer', on='Date', suffixes=('_IEMG', '_ACWV'))
d8 = pd.merge(d7, dfVSS, how='outer', on='Date', suffixes=('_ACWV', '_VSS'))
d9 = pd.merge(d8, dfVB, how='outer', on='Date', suffixes=('_VSS', '_VB'))
d10 = pd.merge(d9, dfNFRA, how='outer', on='Date', suffixes=('_VB', '_NFRA'))
d11 = pd.merge(d10, dfIYW, how='outer', on='Date', suffixes=('_NFRA', '_IYW'))
d12 = pd.merge(d11, dfUSRT, how='outer', on='Date', suffixes=('_IYW', '_USRT'))
d13 = pd.merge(d12, dfIAGG, how='outer', on='Date', suffixes=('_USRT', '_IAGG'))
d14 = pd.merge(d13, dfEMB, how='outer', on='Date', suffixes=('_IAGG', '_EMB'))
d15 = pd.merge(d14, dfHEFA, how='outer', on='Date', suffixes=('_EMB', '_HEFA'))
d16 = pd.merge(d15, dfHEEM, how='outer', on='Date', suffixes=('_HEFA', '_HEEM'))
d17 = pd.merge(d16, dfGAL, how='outer', on='Date', suffixes=('_HEEM', '_GAL'))
```

```python
d17
```

```python
df = d17
```

```python
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr()

plt.figure(figsize = (30,15))
sns.heatmap(correlation_matrix, annot = True)

plt.title("Corr. Matrix My Assets 2015-2022")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show
```
