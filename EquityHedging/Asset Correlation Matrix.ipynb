```python
import yfinance as yF
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
stock_list = ['EEM', 'EUM']
print('stock_list:', stock_list)
```

```python
data = yF.download(stock_list, start="2020-01-01", end="2022-06-18")
```

```python
data.drop(['Adj Close','Volume','High','Open','Low'], axis=1, inplace=True)
```

```python
data.columns = data.columns.map(''.join).str.strip('')
```

```python
data.columns = data.columns.str.lstrip("Close")
```

```python
data
```

```python
data.corr(method='pearson')
```

```python
x_axis_labels = [1,2]
y_axis_labels = [1,2]
```

```python
correlation_matrix = data.corr()

sns.heatmap(correlation_matrix, annot=True, xticklabels=x_axis_labels, yticklabels=y_axis_labels)
plt.title('Em & Inverse Hedge Correlation')
plt.savefig('Em.jpg')
```
