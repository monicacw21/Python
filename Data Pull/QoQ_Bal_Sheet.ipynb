```python
import yfinance as yF
import os  
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
stock_data = yF.Ticker("AAPL")
```

```python
q_fin = stock_data.quarterly_balance_sheet
```

```python
q_fin = q_fin.transpose()
```

```python
q_fin = q_fin.rename_axis("date")
```

```python
q_fin = q_fin.sort_values(by='date')
```

```python
q_fin
```

```python
data = q_fin
```

```python
plt.figure(figsize = (20,12))

sns.lineplot(data=data)
plt.title('Balance Sheet QoQ Change', size=25)
plt.xlabel('Date', size=15)
plt.ylabel('Change in $', size=15)
```
