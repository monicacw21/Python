```python
import yfinance as yF
import os  
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
stock_data = yF.Ticker("AAPL")
```

```python
y_fin = stock_data.financials
```

```python
y_fin = y_fin.transpose()
```

```python
y_fin = y_fin.rename_axis("date")
```

```python
y_fin = y_fin.sort_values(by='date')
```

```python
y_fin
```

```python
data = y_fin
```

```python
plt.figure(figsize = (20,12))

sns.lineplot(data=data)
plt.title('Income Statement YoY Change', size=25)
plt.xlabel('Date', size=15)
plt.ylabel('Change in $', size=15)
```
