```python
import yfinance as yF
import os  
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
stock_data = yF.Ticker("AAPL")
```

```python
q_cas = stock_data.quarterly_cashflow
```

```python
q_cas = q_cas.transpose()
```

```python
q_cas = q_cas.rename_axis("date")
```

```python
q_cas = q_cas.sort_values(by='date')
```

```python
q_cas
```

```python
data = q_cas
```

```python
plt.figure(figsize = (20,12))

sns.lineplot(data=data)
plt.title('Cashflow QoQ Change', size=25)
plt.xlabel('Date', size=15)
plt.ylabel('Change in $', size=15)
```
