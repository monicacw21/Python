```python
import yfinance as yF
import os  
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
stock_data = yF.Ticker("AAPL")
```

```python
y_cas = stock_data.cashflow
```

```python
y_cas = y_cas.transpose()
```

```python
y_cas = y_cas.rename_axis("date")
```

```python
y_cas = y_cas.sort_values(by='date')
```

```python
y_cas
```

```python
data = y_cas
```

```python
plt.figure(figsize = (20,12))

sns.lineplot(data=data)
plt.title('Cashflow YoY Change', size=25)
plt.xlabel('Date', size=15)
plt.ylabel('Change in $', size=15)
```
