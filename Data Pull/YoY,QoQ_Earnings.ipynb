```python
import yfinance as yF
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
stock_data = yF.Ticker("TAL")
```

```python
data_quarterly = stock_data.quarterly_earnings
```

```python
data_quarterly
```

```python
plt.figure(figsize = (10,8))

sns.lineplot(data=data_quarterly)
plt.title('Revenue QoQ Change', size=25)
plt.xlabel('Date', size=15)
plt.ylabel('Change in $', size=15)
```

```python
data_year = stock_data.earnings
```

```python
data_year
```

```python
plt.figure(figsize = (10,8))

sns.lineplot(data=data_year)
plt.title('Earnings YoY Change', size=25)
plt.xlabel('Date', size=15)
plt.ylabel('Change in $', size=15)
```
