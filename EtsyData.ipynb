```python
plt.figure(figsize = (15,8))
sns.regplot(x='sold_count', y='total_rating_count', data=df)
```

```python
correlation_matrix = df.corr()

plt.figure(figsize = (15,8))
sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation Matrix Etsy Sellers")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show
```
