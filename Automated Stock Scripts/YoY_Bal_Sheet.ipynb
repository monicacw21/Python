```python
import yfinance as yF
import os  
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
stock_data = yF.Ticker("AAPL")
```

```python
y_fin = stock_data.balance_sheet
```

```python
y_fin = y_fin.transpose()
```

```python
y_fin = y_fin.rename_axis("date")
```

```python
y_fin = y_fin.sort_values(by='date')
```

```python
y_fin
```

```python
data = y_fin
```

```python
yearly_bal_change = data.pct_change(1).mul(100)
```

```python
yearly_bal_change = yearly_bal_change.round(2)
```

```python
yearly_bal_change
```

```python
yearly_bal_change = yearly_bal_change.iloc[1: , :]
```

```python
yearly_bal_change = yearly_bal_change.dropna(axis='columns')
```

```python
yearly_bal_change
```

```python
plt.figure(figsize = (30,12))
sns.heatmap(yearly_bal_change, annot = True, cmap="YlGnBu")


plt.title("Balance Sheet Change YoY %", size=20)
plt.xlabel("Feature")
plt.ylabel("Year")
plt.show
```
