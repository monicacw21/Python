```python
import yfinance as yF
import os  
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
stock_data = yF.Ticker("AAPL")
```

```python
q_cas = stock_data.quarterly_cashflow
```

```python
q_cas = q_cas.transpose()
```

```python
q_cas = q_cas.rename_axis("date")
```

```python
q_cas = q_cas.sort_values(by='date')
```

```python
q_cas
```

```python
data = q_cas
```

```python
quarterly_cas_change = data.pct_change(1).mul(100)
```

```python
quarterly_cas_change = quarterly_cas_change.round(2)
```

```python
quarterly_cas_change
```

```python
quarterly_cas_change = quarterly_cas_change.iloc[1: , :]
```

```python
quarterly_cas_change = quarterly_cas_change.dropna(axis='columns')
```

```python
quarterly_cas_change
```

```python
plt.figure(figsize = (30,12))
sns.heatmap(quarterly_cas_change, annot = True, cmap="YlGnBu")


plt.title("Cash Flow Change QoQ %", size=20)
plt.xlabel("Feature")
plt.ylabel("Year")
plt.show
```
