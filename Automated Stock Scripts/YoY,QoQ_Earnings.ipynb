```python
import yfinance as yF
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
stock_data = yF.Ticker("TAL")
```

```python
data_quarterly = stock_data.quarterly_earnings
```

```python
quarterly_change = data_quarterly.pct_change(1).mul(100)
```

```python
data_quarterly
```

```python
quarterly_change
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(quarterly_change, annot = True, cmap="YlGnBu")


plt.title("Earnings Change QoQ %", size=20)
plt.xlabel("Features")
plt.ylabel("Features")
plt.show
```
