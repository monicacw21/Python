```python
import yfinance as yF
import os  
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
stock_data = yF.Ticker("AAPL")
```

```python
q_fin = stock_data.quarterly_balance_sheet
```

```python
q_fin = q_fin.transpose()
```

```python
q_fin = q_fin.rename_axis("date")
```

```python
q_fin = q_fin.sort_values(by='date')
```

```python
q_fin
```

```python
data = q_fin
```

```python
quarterly_bal_change = data.pct_change(1).mul(100)
```

```python
quarterly_bal_change = quarterly_bal_change.round(2)
```

```python
quarterly_bal_change
```

```python
quarterly_bal_change = quarterly_bal_change.iloc[1: , :]
```

```python
quarterly_bal_change = quarterly_bal_change.dropna(axis='columns')
```

```python
quarterly_bal_change
```

```python
plt.figure(figsize = (30,12))
sns.heatmap(quarterly_bal_change, annot = True, cmap="YlGnBu")


plt.title("Balance Sheet Change QoQ %", size=20)
plt.xlabel("Feature")
plt.ylabel("Year")
plt.show
```
